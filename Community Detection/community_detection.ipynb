```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
from networkx.algorithms.community import greedy_modularity_communities, asyn_lpa_communities
```

```python
# Load Dataset - Zachary's Karate Club Graph
G = nx.karate_club_graph()
```

```python
# Louvain Method
louvain_communities = list(greedy_modularity_communities(G))
```

```python
# Girvan-Newman Algorithm
comp = community.girvan_newman(G)
girvan_newman_communities = tuple(sorted(c) for c in next(comp))
```

```python
# Label Propagation Algorithm
label_propagation_communities = list(asyn_lpa_communities(G))
```

```python
# visualize detected communities
def draw_communities(G, communities, title):
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G)
    colors = [i for i, comm in enumerate(communities) for node in comm]
    nodes = [node for comm in communities for node in comm]
    nx.draw(G, pos, node_color=colors, cmap=plt.cm.Set3, with_labels=True, node_size=500, alpha=0.8)
    plt.title(title)
    plt.show()
```

```python
draw_communities(G, louvain_communities, "Louvain Community Detection")
```

```python
draw_communities(G, girvan_newman_communities, "Girvan-Newman Community Detection")
```

```python
draw_communities(G, label_propagation_communities, "Label Propagation Community Detection")
```

```python
# Print Community Results
print("Louvain Communities:")
print([list(comm) for comm in louvain_communities])
```

```python
print("Girvan-Newman Communities:")
print([list(comm) for comm in girvan_newman_communities])
```
