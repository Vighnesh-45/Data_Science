```python
import pandas as pd

# Sample product sales data
data = [
    ["ProductID", "ProductName", "Category", "Price", "QuantitySold", "Region"],
    [101, "Laptop", "Electronics", 900, 150, "North America"],
    [102, "Smartphone", "Electronics", 700, 300, "Europe"],
    [103, "Tablet", "Electronics", 400, 250, "Asia"],
    [104, "Office Chair", "Furniture", 120, 500, "North America"],
    [105, "Desk Lamp", "Furniture", 45, 600, "Europe"],
    [106, "Coffee Maker", "Appliances", 80, 200, "North America"],
    [107, "Blender", "Appliances", 60, 450, "Asia"],
    [108, "Microwave", "Appliances", 100, 350, "Europe"]
]

# Save data as CSV
df = pd.DataFrame(data[1:], columns=data[0])
df.to_csv('Product_sales_data.csv', index=False)
df
```

```python
# view sales data
!cat /content/Product_sales_data.csv
```

```python
# replace "North America" with "United States" in the Country Column
!sed 's/North America/United States/g' /content/Product_sales_data.csv
```

```python
# remove all rows where the department is "Tablet":
!sed '/Tablet/d' /content/Product_sales_data.csv
```

```python
# extract only the Product ID and Category
!gawk -F, '{print $1 ", " $3}' /content/Product_sales_data.csv
```

```python
# filter rows where price is greater Than 100
!gawk -F, '$4 > 100 {print $0}' /content/Product_sales_data.csv
```

```python
# Average Quantity sold
!gawk -F, 'NR > 1 {sum+=$5; count++} END {print "Average Quanity Sold: " sum/count}' /content/Product_sales_data.csv
```
