```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import make_blobs
import pandas as pd
```

```python
# Load dataset
dataset_file = "/content/sample_data/2D_dataset.csv"
df = pd.read_csv(dataset_file)
X = df.values
```

```python
# Perform hierarchical clustering with different linkage methods
linkage_methods = ['single', 'complete', 'average']
```

```python
# Plot the dataset
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c='blue', edgecolors='k')
plt.title("Generated 2D Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Create and plot dendrograms for different linkage methods
fig, axes = plt.subplots(1, 3, figsize=(25, 10))

for i, method in enumerate(linkage_methods):
    Z = linkage(X, method=method)
    axes[i].set_title(f"Dendrogram ({method} linkage)")
    dendrogram(Z, ax=axes[i])

plt.tight_layout()
plt.show()
```
